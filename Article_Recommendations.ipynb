```python
data = data[['text', 'title']]
```

```python
data= data.dropna()
```

```python
data= data[:40]
print(data)
```

```python
news = data["text"].tolist()
uni_tfidf = text.TfidfVectorizer(input=news, stop_words="english")
uni_matrix = uni_tfidf.fit_transform(news)
uni_sim = cosine_similarity(uni_matrix)
def recommend(x):
    return ", ".join(data["title"].loc[x.argsort()[:1]])    
data["recommended articles"] = [recommend(x) for x in uni_sim]
```

```python
recommend_articles = data[['title', 'recommended articles']]
recommend_articles
```
